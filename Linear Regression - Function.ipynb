```python
import numpy as np
import matplotlib.pyplot as plt

# function(array of x features, array of y features, number of iterations)
def linear_regression(x_array, y_array, iteration):
  teta = np.random.rand()*10 #Random Number
  b = 1 #Random Number
  teta_MSE = teta
  b_MSE = b

  iteration = iteration
  learning_rate = 0.1

  #MAE
  for i in range(iteration):
    y_hat = teta * x_array + b
    teta = teta - learning_rate * 1/3 * np.sum(np.dot(np.sign(y_hat - y_array), x_array))
    b = b - learning_rate * 1/3 * np.sum(np.sign(y_hat - y_array))
  #MSE
  for i in range(iteration):
    y_hat_MSE = teta_MSE * x_array + b_MSE
    teta_MSE = teta_MSE - learning_rate * 1/3 * np.sum(np.dot(y_hat_MSE - y_array, x_array))
    b_MSE = b_MSE - learning_rate * 1/3 * np.sum(y_hat_MSE - y_array)

  return teta, b, teta_MSE, b_MSE

#Random Array
x_array = np.array([1.2,2.5,3.6])
y_array = np.array([8.5,9.2,10.1])

reg = linear_regression(x_array, y_array, 1000)

# Plot MAE
plt.scatter(x_array, y_array, color='blue', marker='o', label='Y')
plt.plot(x_array, reg[0] * x_array + reg[1], color='red')
plt.xlabel('X')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.title('MAE')
plt.show()

# Plot MSE
plt.scatter(x_array, y_array, color='blue', marker='o', label='Y')
plt.plot(x_array, reg[2] * x_array + reg[3], color='red')
plt.xlabel('X')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.title('MSE')
plt.show()
```
