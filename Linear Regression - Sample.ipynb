```python
import numpy as np
import matplotlib.pyplot as plt

teta = np.random.rand()*10
b = 1
learning_rate = 0.1

xArray = np.array([1.2,2.5,3.6])

y_hat_0 = teta * xArray + b

yArray = np.array([8.5,9.2,10.1])
```

```python
print(xArray)
print(y_hat_0)
print(yArray)
```

```python
#Initial Plot
plt.scatter(xArray, yArray, color='blue', marker='o', label='Y')
plt.plot(xArray, teta * xArray + b, color='red', label='it 0')
plt.xlabel('X')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.title('MAE')
plt.show()
```

```python
#First Iteration
#MAE
teta_1 = teta - learning_rate *1/3* np.sum(np.dot(np.sign(y_hat_0 - yArray), xArray))
b_1 = b - learning_rate * 1/3 * np.sum(np.sign(y_hat_0 - yArray))
y_hat_1 = teta_1 * xArray + b_1
plt.scatter(xArray, yArray, color='blue', marker='o', label='Y')
plt.plot(xArray, teta * xArray + b, color='red', label='it 0')
plt.plot(xArray, teta_1 * xArray + b_1, color='green', label='it 1')
plt.xlabel('X')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.title('MAE')
plt.show()

#MSE
teta_MSE_1 = teta - learning_rate *1/3* np.sum(np.dot(y_hat_0 - yArray, xArray))
b_MSE_1 = b - learning_rate * 1/3 * np.sum((y_hat_0 - yArray))
y_hat_MSE_1 = teta_MSE_1 * xArray + b_MSE_1
plt.scatter(xArray, yArray, color='blue', marker='o', label='Y')
plt.plot(xArray, teta * xArray + b, color='red', label='it 0')
plt.plot(xArray, teta_MSE_1 * xArray + b_MSE_1, color='green', label='it 1')
plt.xlabel('X')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.title('MSE')
plt.show()
```

```python
#Second Iteration
#MAE
teta_2 = teta_1 - learning_rate *1/3* np.sum(np.dot(np.sign(y_hat_1 - yArray), xArray))
b_2 = b_1 - learning_rate * 1/3 * np.sum(np.sign(y_hat_1 - yArray))
y_hat_2 = teta_2 * xArray + b_2
plt.scatter(xArray, yArray, color='blue', marker='o', label='Y')
plt.plot(xArray, teta * xArray + b, color='red', label='it 0')
plt.plot(xArray, teta_1 * xArray + b_1, color='green', label='it 1')
plt.plot(xArray, teta_2 * xArray + b_2, color='yellow', label='it 2')
plt.xlabel('X')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.title('MAE')
plt.show()

#MSE
teta_MSE_2 = teta_MSE_1 - learning_rate *1/3* np.sum(np.dot(y_hat_MSE_1 - yArray, xArray))
b_MSE_2 = b_MSE_1 - learning_rate * 1/3 * np.sum((y_hat_MSE_1 - yArray))
y_hat_MSE_2 = teta_MSE_2 * xArray + b_MSE_2
plt.scatter(xArray, yArray, color='blue', marker='o', label='Y')
plt.plot(xArray, teta * xArray + b, color='red', label='it 0')
plt.plot(xArray, teta_MSE_1 * xArray + b_MSE_1, color='green', label='it 1')
plt.plot(xArray, teta_MSE_2 * xArray + b_MSE_2, color='yellow', label='it 2')
plt.xlabel('X')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.title('MSE')
plt.show()
```

```python
#Second Iteration
#MAE
teta_3 = teta_2 - learning_rate *1/3* np.sum(np.dot(np.sign(y_hat_2 - yArray), xArray))
b_3 = b_2 - learning_rate * 1/3 * np.sum(np.sign(y_hat_2 - yArray))
y_hat_3 = teta_3 * xArray + b_3
plt.scatter(xArray, yArray, color='blue', marker='o', label='Y')
plt.plot(xArray, teta * xArray + b, color='red', label='it 0')
plt.plot(xArray, teta_1 * xArray + b_1, color='green', label='it 1')
plt.plot(xArray, teta_2 * xArray + b_2, color='yellow', label='it 2')
plt.plot(xArray, teta_3 * xArray + b_2, color='black', label='it 3')
plt.xlabel('X')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.title('MAE')
plt.show()

#MSE
teta_MSE_3 = teta_MSE_2 - learning_rate *1/3* np.sum(np.dot(y_hat_MSE_2 - yArray, xArray))
b_MSE_3 = b_MSE_2 - learning_rate * 1/3 * np.sum((y_hat_MSE_2 - yArray))
y_hat_MSE_3 = teta_MSE_3 * xArray + b_MSE_3
plt.scatter(xArray, yArray, color='blue', marker='o', label='Y')
plt.plot(xArray, teta * xArray + b, color='red', label='it 0')
plt.plot(xArray, teta_MSE_1 * xArray + b_MSE_1, color='green', label='it 1')
plt.plot(xArray, teta_MSE_2 * xArray + b_MSE_2, color='yellow', label='it 2')
plt.plot(xArray, teta_MSE_3 * xArray + b_MSE_3, color='black', label='it 3')
plt.xlabel('X')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.title('MSE')
plt.show()
```
